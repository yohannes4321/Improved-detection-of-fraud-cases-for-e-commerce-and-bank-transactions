```python
import numpy as np
df['Hour'] = np.floor(df['Time'] / 3600)  
legit_df = df[df['Class'] == 0]
fraud_df = df[df['Class'] == 1]
legit_counts = legit_df.groupby('Hour')['Class'].count()
fraud_counts = fraud_df.groupby('Hour')['Class'].count()
hourly_distribution = pd.DataFrame({
    'Legit': legit_counts,
    'Fraud': fraud_counts
}).fillna(0)  
import matplotlib.pyplot as plt
import seaborn as sns

fig,(axs1,axs2)=plt.subplots(ncols=2,figsize=(15, 8))
sns.lineplot(ax=axs1,data=hourly_distribution, x='Hour',y='Legit')
sns.lineplot(ax=axs2,data=hourly_distribution, x='Hour',y='Fraud',color='red')
axs1.set_title('Number of Non Fraud Transactions by Hour')
axs2.set_title('Number of Fraud Transactions by Hour')

plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
X=final_df.drop(['V22','V23','V24','V25','V26'],axis=1)
Y=final_df['Class']
columns_x=X.columns
from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler()
X_scaled=m.fit_transform(X)
X=pd.DataFrame(X_scaled,columns=columns_x)
X.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
models={
    'Logistric Regression':LogisticRegression(max_iter=1000),
    'RandomForestClassifier':RandomForestClassifier(n_estimators=100,random_state=42,class_weight='balanced')

}
result=[]
for name,model in models.items():
    model.fit(x_train,y_train)
    predict=model.predict(x_test)
    accuracy=accuracy_score(y_test,predict)
    prec=precision_score(y_test,predict)
    rec=recall_score(y_test,predict)
    f1=f1_score(y_test,predict)
    roc_auc=roc_auc_score(y_test,predict)
    result.append((name,accuracy,prec,rec,f1,roc_auc))
data=pd.DataFrame(result,columns=['Model_name','Acuracy','Preceision','Recall','F1-score','ROC AUC SCORE'])
data.head()

    

```
